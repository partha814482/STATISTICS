```python
# Count values by gender and disorder
male_counts = data[data['Gender'] == 'Male']['Sleep Disorder'].value_counts()
female_counts = data[data['Gender'] == 'Female']['Sleep Disorder'].value_counts()

# Align categories
categories = sorted(set(data['Sleep Disorder'].dropna()))
x = range(len(categories))

plt.bar([i - 0.2 for i in x], [male_counts.get(cat, 0) for cat in categories],
        width=0.4, label='Male')
plt.bar([i + 0.2 for i in x], [female_counts.get(cat, 0) for cat in categories],
        width=0.4, label='Female')

plt.xticks(x, categories, rotation=45, ha='right')
plt.xlabel('Sleep Disorder')
plt.ylabel('No. of People')
plt.title('Distribution of Sleep Disorder by Gender')
plt.legend()
plt.tight_layout()
plt.show()
```

##['Sleep Disorder'].value_counts() â†’ counts how many males fall into each sleep disorder category.
##set(...).dropna() â†’ gets unique sleep disorder types (ignores missing values).

##sorted(...) â†’ sorts them alphabetically for consistent x-axis order.

##x = range(len(categories)) â†’ assigns numeric positions (0,1,2,...) for each category.
##[i - 0.2 for i in x] â†’ shifts male bars slightly left.

##[i + 0.2 for i in x] â†’ shifts female bars slightly right.

##width=0.4 â†’ each bar takes 40% width â†’ creates space between them.

##get(cat, 0) â†’ ensures if a category is missing (e.g., no female with "Sleep Apnea"), it returns 0 instead of error.
